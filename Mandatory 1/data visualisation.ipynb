```python
import matplotlib.pyplot as plt
import numpy as np

# Define measurement distances
measurement_distances = [0, 1.5, 3.0, 4.5, 6.0]

# Define sensor data
sensor_data = {
    "Sensor E": [
        [48.0] * 9 + [48.0],
        [11.0] * 9 + [11.65],
        [5.0] * 10,
        [2.0] * 10,
        [1.0] * 10
    ],
    "Sensor F": [
        [53.2, 54.0, 54.0, 54.0, 54.0, 54.0, 54.0, 54.0, 54.0, 54.0],
        [12.0] * 10,
        [5.0] * 10,
        [3.0] * 10,
        [2.0] * 10
    ],
    "Sensor C": [
        [47.0] * 10,
        [11.0] * 10,
        [4.0] * 10,
        [2.0] * 10,
        [1.0] * 10
    ]
}

# Create plots
plt.figure(figsize=(10, 6))

for sensor, values in sensor_data.items():
    mean_values = [np.mean(v) for v in values]  # Compute mean for each distance
    plt.plot(measurement_distances, mean_values, marker='o', label=sensor)

plt.xlabel("Measurement Distance (cm)")
plt.ylabel("Sensor Readings (Mean)")
plt.title("Sensor Measurements vs. Distance")
plt.legend()
plt.grid()
plt.show()
```
